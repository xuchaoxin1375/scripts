```python
import threading
import queue
import time
from datetime import datetime
def get_time(accurate=False):
    if(accurate):
        # res=datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        res=datetime.now().strftime("%H:%M:%S.%f")
    else:
        res=datetime.now().strftime("%H:%M:%S")
    # print(res)
    return res
    
def log(msg):
    print(f"[{get_time()}]{msg}")
```

```python


q = queue.Queue(2)  # 最大容量2

# 定义生产者,time.sleep模拟每0.1秒生产1个任务,并生产共计5个任务后不再生产
def producer():
    log("开始生产,共计5个")
    for i in range(5):
        print(f'[{get_time()}]放入 {i}')
        q.put(i)
        time.sleep(1)
        if q.full():
            log("队列已满,等待消费后腾出位置")
    log("停止生产")

# 定义消费者,利用time.sleep模拟取出队列中的任务处理耗时,当任务处理完毕后执行task_done()
def consumer():
    log("开始消费,直到队列为空结束")
    while True:
        item = q.get()
        print(f'[{get_time()}]取出 {item}')
        time.sleep(3) # 模拟任务处理
        q.task_done()  # 标记任务完成
        if q.empty():
            break
    log("全部任务取出并结束")

# 启动线程(作为守护线程在后台运行)
threading.Thread(target=producer, daemon=True).start()
threading.Thread(target=consumer, daemon=True).start()

q.join()  # 等待所有任务完成
print("所有线程(生产者和消费者)结束")
```
