```python
import os 
import pandas as pd
```

```python
# img_dir=r"D:\wp_templates\aventuraalairelibre.com\wp-content\uploads\2025"
# images=os.listdir(img_dir)
images=os.listdir('.')
```

```python
csv_dir=r"D:\han\导出数据\0605-西班牙户外"
#将所有csv文件合并成一个,保存到df中
dfs=[]
for file in os.listdir(csv_dir):
    if file.endswith('.csv'):
        file_path = os.path.join(csv_dir, file)
        df = pd.read_csv(file_path, encoding='utf-8')
        dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
```

```python
df
```

```python
#将df中的Images字段转换为webp
df['Images'] = df['Images'].str.replace('.jpg', '.webp')
```

```python
df
```

```python
# 如果df中的images字段取值在images中，则将该行提取出来,最终这些行导出为一个csv文件\
df_filtered = df[df['Images'].isin(images)]
# 将筛选后的数据导出为新的CSV文件
output_csv_path = os.path.join(csv_dir, 'camper.csv')
df_filtered.to_csv(output_csv_path, index=False, encoding='utf-8')
```
