```python
"""
图片水印去除工具
==========================
支持格式：JPG, JPEG, PNG, BMP, TIFF 等常见图片格式

原理说明：
- 通过指定水印所在的矩形区域，使用 OpenCV 的 inpaint 算法进行修复。
- 适合去除图片边角、固定区域的水印（如 logo、文字等）。
- 不适合复杂背景或大面积水印。

依赖库：
- opencv-python
- numpy

安装依赖：
    pip install opencv-python numpy

用法示例：
    python remove_watermark.py --input input.jpg --output output.jpg --rect 10 10 100 40

参数说明：
    --input   输入图片路径
    --output  输出图片路径
    --rect    水印区域，格式为 x y w h（左上角坐标和宽高，单位像素）

批量处理示例：
    python remove_watermark.py --input_dir ./imgs --output_dir ./out --rect 10 10 100 40

"""

import cv2
import numpy as np
import os
import argparse

def remove_watermark(img, rect):
    x, y, w, h = rect
    mask = np.zeros(img.shape[:2], np.uint8)
    mask[y:y+h, x:x+w] = 255
    result = cv2.inpaint(img, mask, 3, cv2.INPAINT_TELEA)
    return result

def process_file(input_path, output_path, rect):
    img = cv2.imread(input_path)
    if img is None:
        print(f"无法读取图片: {input_path}")
        return
    result = remove_watermark(img, rect)
    cv2.imwrite(output_path, result)
    print(f"已处理: {input_path} -> {output_path}")

def main():
    parser = argparse.ArgumentParser(description="图片水印去除工具")
    parser.add_argument('--input', help='输入图片路径')
    parser.add_argument('--output', help='输出图片路径')
    parser.add_argument('--input_dir', help='输入文件夹（批量处理）')
    parser.add_argument('--output_dir', help='输出文件夹（批量处理）')
    parser.add_argument('--rect', nargs=4, type=int, required=True, metavar=('x', 'y', 'w', 'h'), help='水印区域: x y w h')
    args = parser.parse_args()

    if args.input and args.output:
        process_file(args.input, args.output, args.rect)
    elif args.input_dir and args.output_dir:
        os.makedirs(args.output_dir, exist_ok=True)
        for fname in os.listdir(args.input_dir):
            if fname.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff')):
                in_path = os.path.join(args.input_dir, fname)
                out_path = os.path.join(args.output_dir, fname)
                process_file(in_path, out_path, args.rect)
    else:
        print("请指定 --input/--output 或 --input_dir/--output_dir")

if __name__ == '__main__':
    main()
```
