```python
import os
import pandas as pd

workdir=r"C:\Users\Administrator\Desktop\data_output\ukleafhouse.com"
files=os.listdir(workdir)
# files
```

```python
for file in files:
    path=os.path.join(workdir,file)
    print(f"filepath{path}")
    df=pd.read_csv(path)
    df['Description']=df['Description'].str.replace(r'Seller Inventory # \d+','',regex=True).str.replace(r'For further info on this title, click on the &#34;Contact Seller&#34','',regex=True)
    df.to_csv(path,index=False)
```
