```python
import cv2
import numpy as np

# 读取图片
image = cv2.imread("C:\\share\\MK\\img.png", cv2.IMREAD_GRAYSCALE)

# 应用Canny边缘检测
edges = cv2.Canny(image, threshold1=100, threshold2=200)

# 显示原图和处理后的图像
cv2.imshow("Original Image", image)
cv2.imshow("Edges using Canny", edges)

# 等待按键后关闭所有窗口
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt

def extract_edges(image_path, threshold1, threshold2):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    blurred = cv2.GaussianBlur(image, (5, 5), 0)
    edges = cv2.Canny(blurred, threshold1, threshold2)
    return edges

# Replace 'your_image_path.jpg' with the path to your image
edges_image = extract_edges("C:\\share\\MK\\img.png", 100, 200)

# Display the edges using matplotlib
plt.figure(figsize=(8, 6))
plt.imshow(edges_image, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')  # Hide axis
plt.show()
```

```python
import cv2
import numpy as np

# 读取图片
image = cv2.imread(path)

# 创建一个更强的锐化滤波器核
sharpening_kernel = np.array([[-1, -1, -1], [-1, 9 + 8, -1], [-1, -1, -1]])

# 应用滤波器核进行锐化
sharpened_image = cv2.filter2D(image, -1, sharpening_kernel)

# 显示锐化后的图片
cv2.imshow("Sharpened Image", sharpened_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
